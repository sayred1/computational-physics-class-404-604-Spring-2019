f(x)=\pi/2+2*(sin(x)+sin(3x)/3+sin(5x/5)+....sin(nx/n))

```python
#this returns the calculation for n = 11, 21, and 31 
def func(n,x):
    sini = np.zeros(x.shape) #zero array same size as -pi to pi interval
    for i in range(n+1): 
        if i % 2 != 0: #accepts only odd values, i.e: 1,3,5,7...,11; 1,3,5,7...,21; 1,3,5,7...,31
            for j in range(len(xi)): #calculates the sums up to 11, 21, and 31
                sini[j] = sini[j] + np.sin(i*x[j])/i 
    return(np.pi*0.5+2*sini) #returns f(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline  

xi = np.arange(-np.pi,np.pi,0.1) #array of x values from -pi to pi
n = [11,21,31] #list of n values

#plot of the three different curves on one graph
plt.figure(figsize=(15,5))
for i in range(len(n)):
    yi = func(n[i],xi) 
    plt.plot(xi,yi,label = 'n = '+str(n[i]))
pies = np.array([r"{}$\pi$".format(round(i/3,2)) for i in range(-3,4)]) 
plt.xticks([i for i in range(-3,4)],pies)
plt.title('f(x)=pi/2+2*(sin(x)+sin(3x)/3+sin(5x/5)+....sin(nx/n))')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(loc='upper left')
plt.show()

from vpython import *

#vpython program animation
oscillation = graph(title='f(x)=pi/2+2*(sin(x)+sin(3x)/3+sin(5x/5)+....sin(nx/n))', xtitle='x', ytitle='f(x)', fast=False, width=800)
colors = [vec(0,0,1),vec(1,0.5,0),vec(0,1,0)]

for i in range(len(n)):
    curves = gcurve(color=colors[i], width=2, label='n='+str(n[i]))
    for j in range(len(xi)):
        rate(10)
        yi = func(n[i],xi)
        curves.plot(xi[j],yi[j])
```
