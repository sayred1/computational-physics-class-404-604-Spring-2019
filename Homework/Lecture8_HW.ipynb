```python
#Homework Class 8

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x, y):
    n = 5
    f=np.zeros(2)
    f[0] = y[1] #inital condition
    f[1] = 1/(1-(x*x)) * ((2*x*y[1])-(n*(n+1)*y[0])) 
    return(f)

def rk4(t, h, y, n):
    k1 = np.zeros(n)
    k2 = np.zeros(n)
    k3 = np.zeros(n)
    k4 = np.zeros(n)
    k1=h*f(t,y)
    k2=h*f(t+0.5*h,y+0.5*k1)
    k3=h*f(t+0.5*h,y+0.5*k2)
    k4=h*f(t+h,y+k3)
    y+=(k1+2.*(k2+k3)+k4)/6.
    return y

dy_min = 1.
dy_max = 2.
x_end = 1.
max_iter = 100
h = 1E-3
goal = 1.

for i in range(max_iter):
    x_arr = []
    y_arr = []
    dy = (dy_min+dy_max)*0.5
    y = [0.,dy] #array of y and y' derivatives 
    #arrays for plotting the legandre polynomials
    x=0
    while(x<goal):
        y = rk4(x,h,y,2) #range-kutta 
        x = x + h 
        x_arr.append(x)
        y_arr.append(y[0])
    #decide which direction to shoot based off of y[0]
    if y[0]<goal:
        dy_min = dy
    else:
        dy_max = dy

plt.title('Legendre polynomials with n=5')
plt.plot(x_arr,y_arr,'k--')
plt.show()
```
