```python
#used the cmath library to calculate for any arbitrary set of coefficiants
import cmath

#first function to calculate the solutions to the quadratic equation
def sol_to_quad1(a,b,c):
    x1 = (-b+cmath.sqrt(b*b-4*a*c))/(2*a)
    x2 = (-b-cmath.sqrt(b*b-4*a*c))/(2*a)
    return x1,x2
#second function to calculate the solutions to the quadratic equation
def sol_to_quad2(a,b,c):
    x1 = (-2*c)/(b+cmath.sqrt(b*b-4*a*c))
    x2 = (-2*c)/(b-cmath.sqrt(b*b-4*a*c))
    return x1,x2 
```

```python
#used the numpy library for specified a,b,c
import numpy as np

#first function to calculate the solutions to the quadratic equation
def sol_to_quad1(a,b,c):
    x1 = (-b+np.sqrt(b*b-4*a*c))/(2*a)
    x2 = (-b-np.sqrt(b*b-4*a*c))/(2*a)
    return x1,x2
#second function to calculate the solutions to the quadratic equation
def sol_to_quad2(a,b,c):
    x1 = (-2*c)/(b+np.sqrt(b*b-4*a*c))
    x2 = (-2*c)/(b-np.sqrt(b*b-4*a*c))
    return x1,x2 
#function to calculate the relative error
def rel_err(realx,calcx):
    return(abs((realx - calcx)/realx))
```

### As shown above, the calculation either rounds or goes infinite for factors less than the specified machine precision. Also, as c gets smaller, there are obvious deviations between the two calculations. The first calculation should be more precise since we divide by a product. In the second calculations, the error grows as c gets smaller since we are dividing by a difference. Both division and difference lead to larger errors.

```python
print('MACHINE PRECISION GOOD UP TO c = 1.00E-16','\n')

for i in range(1,20):
    #analytic solutions using first method
    x1 = sol_to_quad1(1,1,1*10**-i)[0]
    x2 = sol_to_quad1(1,1,1*10**-i)[1]
    #analytic solutions using second method
    x1_prime = sol_to_quad2(1,1,1*10**-i)[0]
    x2_prime = sol_to_quad1(1,1,1*10**-i)[1]
    #precise solutions calculate with numpy
    nump_roots = np.roots([1,1,1*10**-i])
    nump_roots = nump_roots[::-1]
    print('a = %s, b = %s, c = %.2E' % (1,1,1*10**-i))
    print('x1 = %s, x2 = %s' %(x1,x2)) 
    print("x1' = %s', x2' = %s" %(x1_prime,x2_prime))
    print("solutions using numpy: ", nump_roots[0],',',nump_roots[1])
    #relative error calculations between the 4 calculated solutions and numpy's calculated solutions
    while(i<=16):
        print("relative err for x1 = ",rel_err(nump_roots[0],x1))
        print("relative err for x2 = ",rel_err(nump_roots[1],x2))
        print("relative err for x1' = ",rel_err(nump_roots[0],x1_prime))
        print("relative err for x2' = ",rel_err(nump_roots[1],x2_prime))
        if rel_err(nump_roots[0],x1) <= rel_err(nump_roots[0],x1_prime) and rel_err(nump_roots[1],x2) <= rel_err(nump_roots[1],x2_prime):
            print('The first solution to the quadratic equation is more precise','\n')
        else:
            print('The second solution to the quadratic equation is more precise','\n')
        break
```

### My assumption from earlier was correct, the second method to calculate the solutions were much closer to the exact values generated by numpy thus the second method is in general better.
