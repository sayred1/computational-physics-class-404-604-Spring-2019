```python
import numpy as np
#function to calculate the fd and cd derivatives
def deri(y,x,dx,boo):
    if boo == True:
        return (y(x+dx)-y(x))/dx
    else:
        return (y(x+dx/2)-y(x-dx/2))/dx
        
```

# for forward difference, 1

```python
dx = np.array([1*10**-i for i in range(1,17)]) #array of dx values
err = [] #empty list for error calculations
analy = -np.sin(np.pi/4) #exact value using numpy
print('Analytic Result =',analy)
print('Forward Difference Calculation:')
for i in range(len(dx)): #loop through dx calculating the derivative w/ forward difference per dx
    numeric = deri(np.cos,np.pi/4,dx[i],True) 
    err.append(abs(numeric-analy)) #not logged error calc
    print('For dx = %.2E, derivative = %s with an error of %s' % (dx[i], numeric,err[i]))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#plot of forward difference vs dx
plt.figure(figsize=(15,5))
plt.title('Forward Difference Error vs. dx')
plt.xlabel('log10(dx)')
plt.ylabel('log(Error)')
plt.plot(np.log10(dx),np.log10(err),'b-')

print('error ~ dx^[log(err)/log(dx)]:') #approximation for error

#approximation for the error using the above to solve for n
for i in range(len(err)):
    n = np.log(err[i])/np.log(dx[i])
    print('error = %s ~ %s'%(err[i],dx[i]**n))
    
print('optimal value of dx is',dx[err.index(min(err))]) 
```

# for central difference, 2

```python
#same as 2 cells above but for central difference
err2 = [] 
print('Analytic Result =',analy)
print('Central Difference Calculation:')
for i in range(len(dx)):
    numeric = deri(np.cos,np.pi/4,dx[i],False)
    err2.append(abs(numeric-analy))
    print('For dx = %.2E, derivative = %s with an error of %s' % (dx[i], numeric,err[i]))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(15,5))
plt.title('Central Difference Error vs. dx')
plt.xlabel('log10(dx)')
plt.ylabel('log(Error)')
plt.plot(np.log10(dx),np.log10(err2),'k-')
plt.show()

print('error ~ dx^[log(err)/log(dx)]:')

for i in range(len(err2)):
    n = np.log(err2[i])/np.log(dx[i])
    print('error = %s ~ %s'%(err2[i],dx[i]**n))

print('optimal value of dx is',dx[err2.index(min(err2))])
```

## both methods on one plot:

```python
plt.figure(figsize=(15,5))
plt.title('Error vs. dx')
plt.xlabel('log10(dx)')
plt.ylabel('log(Error)')
plt.plot(np.log10(dx),np.log10(err),'b-',label = 'forward difference')
plt.plot(np.log10(dx),np.log10(err2),'k-',label = 'central difference')
plt.axvline(np.log10(dx[err.index(min(err))]),color = 'b',linestyle = '--')
plt.axvline(np.log10(dx[err2.index(min(err2))]),color = 'k',linestyle = '--')
plt.legend(loc='lower left')
plt.show()
```

### central difference is beter since its errors are lower for reasonable values of dx.
